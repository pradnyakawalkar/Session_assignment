```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/train.csv")

df
```

```python
df.head(10)
```

```python
print("Shape:", df.shape)

print("\nInfo:")
df.info()

print("\nDescribe:")
print(df.describe())
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=["int64","float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

print("Numerical Columns")
print(num_cols)

print("\nCategorical Columns")
print(cat_cols)
```

```python


plt.subplot(2,2,1)
sns.histplot(df["Age"])
plt.title("Age Distribution")

plt.subplot(2,2,2)
sns.histplot(df["Fare"])
plt.title("Fare Distribution")

plt.subplot(2,2,3)
sns.histplot(df["SibSp"])
plt.title("SibSp Distribution")

plt.subplot(2,2,4)
sns.histplot(df["Parch"])
plt.title("Parch Distribution")

plt.tight_layout()
```

```python


plt.subplot(1,3,1)
sns.countplot(x="Sex", data=df)
plt.title("Sex")

plt.subplot(1,3,2)
sns.countplot(x="Embarked", data=df)
plt.title("Embarked")

plt.subplot(1,3,3)
sns.countplot(x="Survived", data=df)
plt.title("Survived")

plt.tight_layout()
```

```python


corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
```

```python
print("Average Fare:", df["Fare"].mean())

print("Maximum Fare:", df["Fare"].max())

print("Minimum Fare:", df["Fare"].min())

print("\nAverage Fare by Sex")
print(df.groupby("Sex")["Fare"].mean())
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Sex", y="Fare", data=df)
plt.title("Fare by Sex")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare by Survival")
plt.show()
```
