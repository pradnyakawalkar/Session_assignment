```python
# Q10. Complete Preprocessing Pipeline

from sklearn.preprocessing import StandardScaler

# Load Dataset
data = pd.read_csv("ford_car_dataset.csv")

# Remove Missing Values
data = data.dropna()

# Remove Duplicate Values
data = data.drop_duplicates()

# Input and Output Features
X = data.drop("price", axis=1)
y = data["price"]

# One Hot Encoding
X = pd.get_dummies(X)

# Feature Scaling
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Convert into DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Display Result
print("First 5 Rows of Scaled Data")
print(X_scaled.head())

print("\nTarget Variable")
print(y.head())

# Observation:
# 1. Missing values removed.
# 2. Duplicate rows removed.
# 3. Categorical variables encoded.
# 4. Numeric features scaled.
# 5. Dataset is ready for Machine Learning models.
```
