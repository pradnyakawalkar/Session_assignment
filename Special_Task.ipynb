```python
avg_reading = df.groupby("parental level of education")["reading score"].mean()

plt.bar(avg_reading.index, avg_reading.values)

plt.title("Average Reading Score by Parental Level of Education")

plt.xlabel("Parental Level of Education")

plt.ylabel("Average Reading Score")

plt.xticks(rotation=45)
```

```python

plt.hist(df["math score"], bins=10)
plt.title("Math Score Distribution")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
```

```python
# Count Plot
sns.countplot(x="gender", data=df)
plt.title("Gender Count")
```

```python
# Scatter Plot
plt.scatter(df["math score"], df["reading score"])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Math vs Reading")
```

```python
sns.boxplot(y=df["writing score"])
plt.title("Writing Score")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
```

```python
gender = df["gender"].value_counts()
plt.pie(gender, labels=gender.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
```
