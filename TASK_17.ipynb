```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/insurance.csv")

```

```python
#q1
#Display first 10 rows
print(df.head(10))
```

```python
#Q2
df.shape
df.info()
df.describe()
```

```python
#Q3
df.isnull().sum()
```

```python
#Q4
# Numerical columns
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
print("Numerical Columns:")
numerical_columns

# Categorical columns
categorical_columns = df.select_dtypes(include=["object"]).columns
print("\nCategorical Columns:")
categorical_columns
```

```python
#Q5
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))

# Age
plt.subplot(2,2,1)
sns.histplot(df["age"])
plt.title("Age Distribution")

# BMI
plt.subplot(2,2,2)
sns.histplot(df["bmi"])
plt.title("BMI Distribution")

# Children
plt.subplot(2,2,3)
sns.histplot(df["children"])
plt.title("Children Distribution")

# Expenses (Use "charges" if your dataset has that column)
plt.subplot(2,2,4)
sns.histplot(df["expenses"])
plt.title("Expenses Distribution")

plt.tight_layout()


```

```python
#Q6
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,4))

# Sex
plt.subplot(1,3,1)
sns.countplot(x="sex", data=df)
plt.title("Sex Count")

# Smoker
plt.subplot(1,3,2)
sns.countplot(x="smoker", data=df)
plt.title("Smoker Count")

# Region
plt.subplot(1,3,3)
sns.countplot(x="region", data=df)
plt.title("Region Count")

plt.tight_layout()
plt.show()
```

```python
#Q7
# Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
```

```python
#Q8
# Average Charges
print("Average Charges:")
print(df["expenses"].mean())

# Maximum Charges
print("\nMaximum Charges:")
print(df["expenses"].max())

# Minimum Charges
print("\nMinimum Charges:")
print(df["expenses"].min())

# Average Charges by Smoker
print("\nAverage expenses by Smoker:")
print(df.groupby("smoker")["expenses"].mean())

```

```python
#Q9
# Charges vs Smoker
sns.boxplot(x="smoker", y="expenses", data=df)
plt.title("Charges by Smoker")
```

```python
#Q9
# Charges vs Sex
sns.boxplot(x="sex", y="expenses", data=df)
plt.title("Charges by Sex")
```
