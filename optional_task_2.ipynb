```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/train.csv")
df
```

```python
df.head(10)
```

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

df.drop("Cabin", axis=1, inplace=True)

print(df.isnull().sum())
```

```python
df.describe()
```

```python


plt.subplot(2,2,1)
plt.hist(df["Age"])
plt.title("Age")

plt.subplot(2,2,2)
plt.hist(df["Fare"])
plt.title("Fare")

plt.subplot(2,2,3)
plt.hist(df["SibSp"])
plt.title("SibSp")

plt.subplot(2,2,4)
plt.hist(df["Parch"])
plt.title("Parch")

plt.tight_layout()
plt.show()
```

```python


plt.subplot(2,2,1)
sns.countplot(x="Survived", data=df)
plt.title("Survived")

plt.subplot(2,2,2)
sns.countplot(x="Pclass", data=df)
plt.title("Passenger Class")

plt.subplot(2,2,3)
sns.countplot(x="Sex", data=df)
plt.title("Gender")

plt.subplot(2,2,4)
sns.countplot(x="Embarked", data=df)
plt.title("Embarked")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Age", y="Fare", data=df)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
plt.figure(figsize=(6,6))

df["Survived"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Survival Percentage")
plt.ylabel("")

plt.show()
```
